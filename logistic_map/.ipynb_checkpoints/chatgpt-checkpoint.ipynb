```python
import matplotlib.pyplot as plt
import numpy as np

def logistic_map(r, x, n):
    """Generate values of the logistic map."""
    values = [x]
    for _ in range(n - 1):
        x = r * x * (1 - x)
        values.append(x)
    return values

# Parameters
r_values = [2.5]
x0 = 0.5
iterations = 100

# Plot
plt.figure(figsize=(12, 8))
for r in r_values:
    values = logistic_map(r, x0, iterations)
    plt.plot(range(iterations), values, label=f'r = {r}')

plt.title('Logistic Map for Different r Values')
plt.xlabel('Iteration')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```
